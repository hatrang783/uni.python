```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
import os

sc = Tk()
sc.title("Employee Management")

sv_login = StringVar()

def login():
    if sv_login.get() == "1234":
        #messagebox.showinfo("Success", "Log-in successfully")
        fr_login.pack_forget()
        fr_main.pack()
    else:
        messagebox.showwarning("Fail", "Wrong password")

##################################################################
# Login Frame
fr_login = Frame(sc)

et_login = Entry(fr_login,show="*", textvariable=sv_login)
et_login.pack()

bt_login = Button(fr_login, text="Log-in", command=login)
bt_login.pack()

###################################################################
# Main Frame
fr_main = Frame(sc)

## Insert Frame
class Form:
    def __init__(self,parent,label,column,row):
        self.frame = Frame(parent)
        self.frame.grid(column=column,row=row)
        self.lb = Label(self.frame,text=label,width=10,anchor=W)
        self.lb.grid(column=1,row=1)
        #if type == "Entry":
        self.et = Entry(self.frame)
        self.et.grid(column=2, row=1)
#         elif type == "ComboBox":
#             self.ccb = ttk.Combobox(self.frame)
#             self.ccb.grid(column=2, row=1)
        
    def set_lb_atb(self,**atb):
        self.lb.config(**atb)
        
    def set_et_atb(self, **atb):
        self.et.config(**atb)
        
    def set_ccb_atb(self, **atb):
        self.ccb.config(**atb)
        
fr_insert = Frame(fr_main)

lb_info = Label(fr_insert, text = "Employee's Information")
lb_info.grid(column=0, row=0)

L = [["Name", "Dob", "Gender", "Phone", "Address"],
     ["Email","Doj", "Department", "Designation", "Salary"],
    ]

D_form = {}

for i in range(len(L)):
    for j in range(len(L[i])):
        D_form[L[i][j]] = Form(fr_insert,L[i][j],i, j+1)
        
### Menu Frame
fr_menu = Frame(fr_insert)

bt_save = Button(fr_menu, text="Save")
bt_save.grid(column=1, row=1)

bt_update = Button(fr_menu, text="Upate")
bt_update.grid(column=1, row=2)

bt_delete = Button(fr_menu, text="Delete")
bt_delete.grid(column=1, row=3)

bt_reset = Button(fr_menu, text="Reset")
bt_reset.grid(column=1, row=4)

fr_menu.grid(column=2, row=0, rowspan=6)
fr_insert.pack()

## Search Frame
fr_search = Frame(fr_main)
lb_search = Label(fr_search, text="Search By:")
lb_search.grid(column=0, row=0)

opt = ['Search by ID', 'Search by Name', 'Search by Department']
ccb_search = ttk.Combobox(fr_search, values = opt)
#ccb_search['values'] = ('Search by ID', 'Search by Name', 'Search by Department')
ccb_search.grid(column=1, row=0)

et_search = Entry(fr_search)
et_search.grid(column=2, row=0)

bt_search = Button(fr_search, text="Search")
bt_search.grid(column=3, row=0)

fr_search.pack()

## Display Frame
fr_display = Frame()

fr_display.pack()

#########################################################################
fr_login.pack()
#fr_main.pack()

sc.mainloop()
```
